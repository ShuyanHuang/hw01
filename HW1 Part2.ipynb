## 1. 
a. When the sample size $n$ is extremely large while the number of predictors $p$ is small, it would be better to use a flexible model. Compared to an inflexible model, a flexible model would be able to find more detailed patterns in the data, since a much larger number of sample relative to predictors prevents the model from overfitting.

b. When the number of predictors $p$ is extremely large while the sample size $n$ is small, it would be better to use an inflexible model. With a small sample size, we are exposed to a high risk of overfitting when we use a flexible model, while an inflexible model is more likely to capture the true patterns out of the noise. Even worse, we may not be able to identify all the parameters in a flexible model because there are more parameters than samples.

c. When the relationship between the predictors and response is highly non-linear, it would be better to use a flexible model. In this case, an inflexible model may not be able to describe the complex non-linear pattern in data. While a flexible model might perform better due to its high expressive power.

d. When the variance of the error terms is extremely high, it would be better to use an inflexible model. In this case, a flexible model may end up fitting the error rather than the true pattern.

b. 

Bias decreases as we go from less flexible statistical learning methods towards more flexible ones. Since real-life problems are usually very complicated, we are introducing bias by using simplied model. Using flexible models can allow us to get closer to the real "model", and thus introducing less bias.

However, variance grows as we use more flexible models. A flexible model follows the data very closely, so a small variation in one data point would cause the estimation to change considerably.

Bayes (irreducible) error does not change with the flexibility of the model, since it's introduced by the fact that real-world data can't be perfectly described by any model.

Since test error is the sum of squared bias, variance and irreducible error, there are two opposit forces from bias and variance affecting test error. When the model is relatively inflexible, the force of bias dominates and test error decreases. As the model becomes more and more flexible, the force of variance becomes dominant and test error increases.

Training error is smaller when we use more flexible models since flexible models can fit the train data better.

```python
import numpy as np

np.random.seed(0)
```

```python
x1 = np.random.uniform(0,1,200)
x2 = np.random.uniform(0,1,200)
```

```python
epsilon = np.random.normal(0, 0.5, 200)
y = x1 + x1**2 + x2 + x2**2 + epsilon
```

```python
def odd2p(x):
    return x/(1+x)

odd2p_vec = np.vectorize(odd2p)
p = odd2p_vec(np.exp(y))
```

```python
import matplotlib.pyplot as plt
from numpy import arange
from numpy import meshgrid

delta = 0.01
xrange = np.arange(-0.5, 1, delta)
yrange = np.arange(-0.5, 1, delta)
X1, X2 = np.meshgrid(xrange,yrange)

fig, ax = plt.subplots()
group = np.array(['success' if i>0.5 else 'failure' for i in p])
for g in np.unique(group):
    i = np.where(group == g)
    ax.scatter(x1[i], x2[i], label=g)
ax.legend()
plt.contour(X1, X2, np.exp(X1+X1**2+X2+X2**2), [1])
plt.title('Simulated two classes and the Bayes decision boundary')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.show()
```
